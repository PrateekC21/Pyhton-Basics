 1. What is NumPy, and why is it widely used in Python?
 - NumPy (Numerical Python) is the fundamental package for scientific computing in Python. Its core feature is the powerful N-dimensional array object, or ndarray.

- Performance: NumPy operations are implemented in C and Fortran, making them much faster than using standard Python lists for numerical computations.

- Vectorized Operations: It allows you to perform mathematical operations on entire arrays without writing explicit loops, which is cleaner and faster.

- Memory Efficiency: NumPy arrays take up less memory compared to Python lists.


 2. How does broadcasting work in NumPy?
 - Broadcasting is a powerful mechanism that allows NumPy to perform arithmetic operations on arrays of different shapes. Instead of duplicating the smaller array to match the larger one's shape (which would be inefficient), NumPy "broadcasts" the smaller array's shape across the larger one.

- The rules for broadcasting are:

- If the arrays do not have the same number of dimensions, prepend 1s to the shape of the smaller array until they do.

- Compare the size of each dimension. The arrays are compatible if, for every dimension, either the sizes are equal or one of them is 1.

- The broadcasted array will have a shape where each dimension's size is the maximum of the input arrays' sizes in that dimension.

4. Explain the use of the groupby() method in Pandas?
- The groupby() method is used for splitting data into groups based on some criteria, applying a function to each group independently, and then combining the results into a data structure. This is often referred to as the "Split-Apply-Combine" strategy.

- Split: The data is split into groups based on the values in one or more columns.

- Apply: An aggregate function (like sum(), mean(), count(), etc.) is applied to each group.

- Combine: The results of the function application are combined into a new DataFrame or Series.

13. Why is handling missing data important in Pandas?
- Handling missing data (often represented as NaN, or Not a Number) is critically important for several reasons:

- Algorithm Compatibility: Many machine learning algorithms cannot process data with missing values and will raise an error.

- Statistical Accuracy: Missing values can skew statistical calculations like mean and standard deviation, leading to biased or incorrect conclusions.

- Data Integrity: Leaving missing data unaddressed can reduce the quality and reliability of your analysis and any models built from it.

- Common strategies include imputation (filling in missing values with a mean, median, or more advanced method) or deletion (removing rows or columns with missing data).

15. How does NumPy handle multidimensional arrays?
- NumPy's core object, the ndarray, is designed to handle multidimensional data efficiently.

- Structure: An ndarray is a grid of values of the same type. Its shape is a tuple of integers indicating the size of the array along each dimension, and its ndim attribute gives the number of dimensions.

- Memory Layout: The data is stored in a single, contiguous block of memory. This allows for highly optimized access and processing, as the CPU can access sequential data much faster.

- Axis-based Operations: NumPy functions can operate along specified axes. For instance, in a 2D array, you can sum values along axis=0 (down the columns) or axis=1 (across the rows).

17. Explain the difference between apply() and map() in Pandas
While both are used for transformations, they operate at different levels.

- map(): This is a Series method only. It works element-wise to substitute each value in a Series with another value. It's best used when you have a dictionary or a simple function to map old values to new ones.

- Example:

# Map 'M' to 'Male', 'F' to 'Female'
df['gender'].map({'M': 'Male', 'F': 'Female'})
apply(): This method can be used on both a DataFrame and a Series.

- On a DataFrame, it applies a function along an axis (either to each column or each row).

- On a Series, it can be used like map(), but it's more flexible for applying complex functions.


# Apply a function to find the range of each column
df[['price', 'cost']].apply(lambda x: x.max() - x.min(), axis=0)

18. What are some advanced features of NumPy?

20. What is the role of a pivot table in Pandas?
- A pivot table is a data summarization tool that reshapes or reorganizes data by converting it from a "long" format to a "wide" format. Its role is to create a new view of your data to help you extract insights by grouping and aggregating it differently.

- In Pandas, the pivot_table() function requires you to specify:

- index: The column to be used for the new DataFrame's rows.

- columns: The column whose unique values will become the new DataFrame's columns.

- values: The column whose values will be aggregated and placed in the new grid.

- aggfunc: The aggregation function to apply (e.g., sum, mean, count).

21. Why is NumPy’s array slicing faster than Python’s list slicing?
- The primary reason is that NumPy slicing creates a view, while Python list slicing creates a copy.

- NumPy Array Slicing (View): When you slice a NumPy array, you get a view of the original array. This means no data is copied. The new slice object simply points to the original data in memory. This is extremely fast and memory-efficient. A side effect is that modifying the slice will also modify the original array.

- Python List Slicing (Copy): When you slice a Python list, a new list is created, and the elements are copied over. This process takes time and consumes additional memory, especially for large lists.

4. How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap?

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Create a sample DataFrame with correlated data
np.random.seed(42)
data = np.random.rand(10, 4)
df = pd.DataFrame(data, columns=['A', 'B', 'C', 'D'])
df['E'] = df['A'] * 2 + np.random.randn(10) * 0.2

# Calculate the correlation matrix
correlation_matrix = df.corr()

# Create a heatmap to visualize the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')

# Add a title
plt.title('Correlation Matrix Heatmap')

# Display the plot
plt.show()
```

11. Perform matrix multiplication using NumPy?

```python
import numpy as np

# Create two matrices
# A is a 2x3 matrix
A = np.array([[1, 2, 3],
              [4, 5, 6]])

# B is a 3x2 matrix
B = np.array([[7, 8],
              [9, 10],
              [11, 12]])

# Perform matrix multiplication using the @ operator
# The result will be a 2x2 matrix
C = A @ B

print("Matrix A (2x3):")
print(A)
print("\nMatrix B (3x2):")
print(B)
print("\nResult of matrix multiplication A @ B (2x2):")
print(C)
```

13. Create a 3D scatter plot using Plotly?

```python
import plotly.express as px
import numpy as np

# Generate sample data for 3D plot
np.random.seed(42)
num_points = 100
x = np.random.randn(num_points)
y = np.random.randn(num_points)
z = np.random.randn(num_points)
# Add a variable for color mapping
color_var = np.sqrt(x**2 + y**2 + z**2)

# Create an interactive 3D scatter plot
fig = px.scatter_3d(
    x=x,
    y=y,
    z=z,
    color=color_var,
    title='Interactive 3D Scatter Plot',
    labels={'x': 'X-Axis', 'y': 'Y-Axis', 'z': 'Z-Axis', 'color': 'Magnitude'}
)

# Customize marker style
fig.update_traces(marker=dict(size=5, opacity=0.8))

# Show the plot
fig.show()
```
