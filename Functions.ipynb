 1. What is the difference between a function and a method in Python?
- Function
  - A function is a block of reusable code that performs a specific task.
  - It is defined independently using the def keyword outside of a class.
  - It is not associated with any object or class.
  - You call a function directly by its name.
  - It does not require a special first parameter like self.
- Method
  - A method is also a function, but it is defined inside a class.
  - It is associated with an object or class.
  - It is called using dot notation on an object.
  - The first parameter is usually self (referring to the instance of the class).
  - Methods can be of three types: instance methods, class methods, and static methods.

 1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in
the list.

```python
def sum_of_even_numbers(numbers):
    even_sum = 0
    for num in numbers:
        if num % 2 == 0:
            even_sum += num
    return even_sum
nums = [47, 11, 42, 13, 20, 8]
result = sum_of_even_numbers(nums)
print("Sum of even numbers:", result)
```

 2. Create a Python function that accepts a string and returns the reverse of that string.

```python
def reverse_string(input_string):
    reversed_string = ""
    for char in input_string:
        reversed_string = char + reversed_string
    return reversed_string
input_str = "Hello, World!"
reversed_str = reverse_string(input_str)
print("Reversed string:", reversed_str)
```

5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of
terms.

```python
class Fibonacci:
    def __init__(self, limit):
        if not isinstance(limit, int) or limit < 0:
            raise ValueError("Limit must be a non-negative integer.")
        self.limit = limit
        self.counter = 0
        self.a, self.b = 0, 1

    def __iter__(self):
        return self
    def __next__(self):
        if self.counter < self.limit:
            current_value = self.a
            self.a, self.b = self.b, self.a + self.b
            self.counter += 1
            return current_value
        else:
            raise StopIteration
if __name__ == "__main__":
    try:
        num_terms = int(input("Enter the number of Fibonacci terms to generate: "))
        fib_iterator = Fibonacci(num_terms)

        print(f"Generating the first {num_terms} terms of the Fibonacci sequence:")
        for number in fib_iterator:
            print(number, end=" ")
        print("\n\nDemonstrating manual iteration with next():")
    except ValueError as e:
        print(f"Error: {e}")
    except StopIteration:
        print("\nIteration stopped as expected.")
```

 6. Write a generator function in Python that yields the powers of 2 up to a given exponent.

```python
def powers_of_two_gen(max_exponent):
    if not isinstance(max_exponent, int) or max_exponent < 0:
        raise ValueError("The maximum exponent must be a non-negative integer.")
    for exponent in range(max_exponent + 1):
        yield 2 ** exponent
if __name__ == "__main__":
    try:
        max_exp = int(input("Enter the maximum exponent for powers of 2: "))
        print(f"Generating powers of 2 up to 2^{max_exp}:")
        for power in powers_of_two_gen(max_exp):
            print(power, end=" ")
        print("\n\nDemonstrating with a list comprehension:")
        powers_list = [p for p in powers_of_two_gen(max_exp)]
        print(powers_list)
    except ValueError as e:
        print(f"Error: {e}")
    except StopIteration:
        print("\nGenerator exhausted as expected.")
```

 7. Implement a generator function that reads a file line by line and yields each line as a string.

```python
def read_file_line_by_line(file_path):
    try:
        with open(file_path, 'r') as file:
            for line in file:
                yield line.strip()
    except FileNotFoundError:
        print("Error: File not found.")
    except Exception as e:
        print(f"An error occurred: {e}")
file_path = "sample.txt"

for line in read_file_line_by_line(file_path):
    print(line)
```

9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.

```python
# List of temperatures in Celsius
celsius = [0, 20, 37, 98, 100, 104]
# Formula: F = (C * 9/5) + 32
fahrenheit = list(map(lambda c: (c * 9/5) + 32, celsius))

print("Temperatures in Celsius   :", celsius)
print("Temperatures in Fahrenheit:", fahrenheit)
```

 10. Create a Python program that uses `filter()` to remove all the vowels from a given string

```python
def remove_vowels(input_string):
    vowels = "aeiouAEIOU"
    filtered_string = ""
    for char in input_string:
        if char not in vowels:
            filtered_string += char
    return filtered_string
#Example
a = "hello sir!"
print(remove_vowels(a))
```
